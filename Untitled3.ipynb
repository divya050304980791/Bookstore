```python
# Install required libraries (run this cell once)
!pip install gradio gTTS playsound SpeechRecognition pydub

# ------------------- AI College Chatbot -------------------
import gradio as gr
from gtts import gTTS
import speech_recognition as sr
import os

# ------------------- College Info -------------------
college_info = {
    "name": "JSS Academy of Technical Education, Noida",
    "hods": {
        "cse": "Dr. Kakoli Banerjee",
        "it": "Dr. Luv Kush Sharma",
        "ece": "Dr. Rajeev Tripathi",
        "eee": "Dr. Anuj Jain"
    },
    "placements": {
        "cse": "Average package in CSE is around 6â€“7 LPA, highest ~45 LPA (2024 batch).",
        "it": "Average package in IT is around 5â€“6 LPA.",
        "ece": "ECE students average around 4â€“5 LPA with core + software mix."
    },
    "about": "JSSATE Noida is affiliated to AKTU, Lucknow and approved by AICTE. It offers B.Tech, M.Tech, and MCA programs."
}

# ------------------- Chatbot Logic -------------------
def chatbot_response(text):
    text = text.lower()
    if "hod" in text and "cse" in text:
        return f"The HOD of CSE Department is {college_info['hods']['cse']}."
    elif "hod" in text and "it" in text:
        return f"The HOD of IT Department is {college_info['hods']['it']}."
    elif "placement" in text and "cse" in text:
        return college_info['placements']['cse']
    elif "placement" in text and "it" in text:
        return college_info['placements']['it']
    elif "placement" in text and "ece" in text:
        return college_info['placements']['ece']
    elif "about" in text or "information" in text or "college" in text:
        return college_info['about']
    elif "name" in text:
        return f"This is {college_info['name']}."
    else:
        return "I'm not sure about that. Please ask about HODs, placements, or college info."

# ------------------- Speech Recognition -------------------
def recognize_speech(audio_path):
    recognizer = sr.Recognizer()
    try:
        with sr.AudioFile(audio_path) as source:
            audio_data = recognizer.record(source)
            text = recognizer.recognize_google(audio_data)
            return text
    except Exception as e:
        return f"Error recognizing speech: {e}"

# ------------------- Combined Bot Function -------------------
def chatbot_interface(audio, text_input):
    if audio is not None:
        text = recognize_speech(audio)
    else:
        text = text_input

    reply = chatbot_response(text)

    # Convert reply to voice
    tts = gTTS(reply)
    tts.save("voice_reply.mp3")

    return reply, "voice_reply.mp3"

# ------------------- Gradio Interface -------------------
iface = gr.Interface(
    fn=chatbot_interface,
    inputs=[
        gr.Audio(sources=["microphone"], type="filepath", label="Ask by speaking ðŸŽ¤"),
        gr.Textbox(label="Or type your question ðŸ’¬")
    ],
    outputs=[
        gr.Textbox(label="Chatbot Response"),
        gr.Audio(label="Voice Reply")
    ],
    title="ðŸŽ“ AI College Chatbot - JSSATE Noida",
    description="Ask questions about departments, HODs, and placements!"
)

iface.launch()
```
